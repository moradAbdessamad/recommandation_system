```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Data/movies_data.csv')
df.head()
```

```python
def create_textual_representation(row):
    textual_representation = f"""
    Types : {row['genres']}
    Title : {row['title']}
    Director : {row['director']}
    Cast : {row['actors']}
    Realeased : {row['release_date']}
    Country : {row['production_countries']}
    Language : {row['original_language']}
    Description : {row['overview']}
    """
    return textual_representation
```

```python
df['Textual_representation'] = df.apply(create_textual_representation, axis=1)
df.sample(5)
```

```python
df.to_csv('../Data/movies_data_with_textual_representation.csv', index=False)
```

```python
import faiss #type: ignore
import requests
import numpy as np
import json
import os

save_path = "faiss_index"
progress_file = f"{save_path}_progress.json"
dim = 4096

# Load or create index
if os.path.exists(save_path):
    index = faiss.read_index(save_path)
    with open(progress_file, 'r') as f:
        processed_count = json.load(f)['processed']
else:
    index = faiss.IndexFlatL2(dim)
    processed_count = 0

def process_batch(start_idx, batch_size=100):
    global processed_count
    
    end_idx = min(start_idx + batch_size, len(df['Textual_representation']))
    print(f"Processing batch {start_idx} to {end_idx} of {len(df['Textual_representation'])}")
    
    for i in range(start_idx, end_idx):
        print(f"Processing {i}/{len(df['Textual_representation'])}")
        
        res = requests.post(
            'http://localhost:11434/api/embeddings',
            json={
                'model': 'llama2',
                'prompt': df['Textual_representation'].iloc[i],
                "device": "cuda",
            })
        
        embedding = np.array(res.json()['embedding'], dtype=np.float32).reshape(1, -1)
        index.add(embedding)
        
        processed_count = i + 1
        
    faiss.write_index(index, save_path)
    with open(progress_file, 'w') as f:
        json.dump({'processed': processed_count}, f)
    
    return processed_count

while processed_count < len(df['Textual_representation']):
    processed_count = process_batch(processed_count)
```
